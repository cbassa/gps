# Looking For A Satellite With Cross Correlation

We want to find a sat.

First read in data

```python
from gps import data
from gps import aquire
from numpy import argmax
# read in 10 milliseconds of data assuming the spain data format
init = data.read('../../data/2013_04_04_GNSS_SIGNAL_at_CTTC_SPAIN.dat', data.SPAIN, 0.010)
print "read in", len(init), "samples"
```

Pick a satellite ID and doppler shift to test, then get cross-correlation for that case

```python
sats = []
for i in range(200):
    sats.append(aquire.test(data.SPAIN['sample-rate'], init, 7, i*100 - 10000))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'font.size': 12})


fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(111)
ax.imshow(sats, interpolation='nearest', cmap='afmhot', aspect=40000/200)
plt.yticks([0, 25, 50, 75, 100, 125, 150, 175, 200],
           ["-10 kHz", "-7.5 kHz", "-5 kHz", "-2.5 kHz", "0", "2.5 kHz", "5 kHz", "7.5 kHz", "10 kHz"])
plt.show()


test = """
for i, sat in enumerate(sats):
    fig = plt.figure(figsize=(16,9))
    ax = fig.add_subplot(111)
    plt.title(str(i+1))
    sat = sat[0:4000]
    ax.plot(range(len(sat)), sat)
    plt.show()
"""
```

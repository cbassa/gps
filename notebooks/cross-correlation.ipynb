# Looking For A Satellite With Cross Correlation

We want to find a sat.

First read in data

```python
from gps import data
from gps import aquire
from numpy import argmax, absolute
# read in 10 milliseconds of data assuming the spain data format
init = data.read('../../data/2013_04_04_GNSS_SIGNAL_at_CTTC_SPAIN.dat', data.SPAIN, 0.010)
print "read in", len(init), "samples"
```

Pick a satellite ID and doppler shift to test, then get cross-correlation for that case

```python
sats = []
ticks = range(0,400,50)
tick_labels = ["%d kHz" % ((i-200)/10) for i in ticks]
for i in range(400):
    test = aquire.test(data.SPAIN['sample-rate'], init, 20, i*100 - 20000)
    sats.append(test)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'font.size': 12})


fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(111)
ax.imshow(sats, interpolation='nearest', cmap='afmhot', aspect=40000/400)
plt.yticks(ticks, tick_labels)
plt.show()


test = """
for i, sat in enumerate(sats):
    fig = plt.figure(figsize=(16,9))
    ax = fig.add_subplot(111)
    plt.title(str(i+1))
    sat = sat[0:4000]
    ax.plot(range(len(sat)), sat)
    plt.show()
"""
```
